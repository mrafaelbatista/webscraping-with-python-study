```python
alunos = []

controle = True

while controle:
    x = int(input("Escolha uma das opções: \n1 - Cadastrar novo aluno \n2 - Excluir aluno \n 3 - Alterar \n 4 listar \n0 - Sair"))
    
    if x == 1:
        print("\n******** CADASTRAR ALUNO ********")
        nome = str(input("Digite o nome do aluno: ")).upper()
        idade = int(input("Digite a idade: "))
        telefone = int(input("qual seu telefone?"))
        dic_aluno = {'nome':nome, 'idade':idade, 'telefone':telefone}
        alunos.append(dic_aluno)

    elif x == 2:
        for i in alunos:
            print(i,alunos.index(i))
        alunos.pop((int(input("digite a posição do aluno que vc gostaria de excluir."))))    
        print("aluno excluído")
        
    elif x == 3:
        print("função indisponpivel no momento tente mais tarde.")
        
    elif x == 4:
        print(alunos)
        
            
    elif x == 0:
        controle = False
        print("Saindo")
    else:
        print("Escolha uma opção válida.")
    
```
